```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
csv_path = 'QB_Completions_Predictive_Model/weekly_data_updated_08_23.csv'

data = pd.read_csv(csv_path)

pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')
pd.set_option('display.max_columns', None)
# pd.reset_option('display.max_columns')

fil_data = data.loc[(data['id'] == 1469) & (data['position'] == 'QB')] 



# fil_data = fil_data.sort_values('completions')
fil_data = fil_data.sort_values('season', ascending=False)


# median = fil_data['offense_snaps'].median()

# print(median)

fil_data

focus_columns = ['id', 'position', 'week', 'season', 'season_type', 'completions', 'passing_tds', 'attempts', 'interceptions', 
                 'sacks', 'passing_air_yards', 'rushing_yards', 'rushing_tds', 'rushing_first_downs', 'comp_percentage', 'offense_snaps',
                 'years_played']

discr_data = fil_data[focus_columns]

discr_data
```

This data shows an overview of Russell Wilson's performance as an NFL quarterback over 10 years from 2012 to 2022.

There are some inconsistencies within this data as is illustrated within the 'week' column which doesn't always present an equally sequential order. This is expected since in the NFL and other American sports, teams are occasionally given bye weeks where they do not have a scheduled game that week and will therefore not play. The other explanation for the inconsistent ordering of these rows could be due to an injury that Russell Wilson suffered which may have resulted in a lack of data for the weeks when he had to undergo surgery and consequentially missed several games that season.

I also noticed that the number of completed passes by Russell did not appear to have any obvious correlation with his passing yards. However, when Wilson successfully completed 24 passes or more, a positive correlation appeared to form.

A potential outlier within this dataset would be the low number of offensive snaps within a game. This likely would have been the direct result of the opposing team's offensive strategy to maintain possession of the ball for as many plays as possible and limit Russell's opportunity to snap the ball each quarter. This would significantly limit most of the quarterback KPIs in this dataset, resulting in anomalous data, especially if most other NFL teams would not have employed the same offensive strategy which based on the overiview, either wasn't the case or was not as effectively executed compared to other teams. To handle this outlier, I should calculate the median value of offensive snaps from Russell's data (64.0) and adjust the data based on a minimum offensive snap value and a maximum snap value around the median, either by using the IQR or by defining a given range arbitrarily adjusted to the median.

```python
csv_path = 'QB_Completions_Predictive_Model/weekly_data_updated_08_23.csv'

data = pd.read_csv(csv_path)

pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')
pd.set_option('display.max_columns', None)
# pd.reset_option('display.max_columns')


fil_data = data.loc[(data['id'] == 1469) & (data['position'] == 'QB')] 

# fil_data = fil_data.sort_values('completions')
fil_data = fil_data.sort_values('season', ascending=False)


Q1 = fil_data['offense_snaps'].quantile(0.25)
Q3 = fil_data['offense_snaps'].quantile(0.75)

lower_bound = Q1
upper_bound = Q3




# Addressing missing values
fil_data['years_played'] = fil_data['years_played'].fillna(11.0)
fil_data = fil_data[(fil_data['offense_snaps'] >= lower_bound) & (fil_data['offense_snaps'] <= upper_bound)]


fil_data

focus_columns = ['id', 'position', 'week', 'season', 'season_type', 'completions', 'passing_tds', 'attempts', 'interceptions', 
                 'sacks', 'passing_air_yards', 'rushing_yards', 'rushing_tds', 'rushing_first_downs', 'comp_percentage', 'offense_snaps',
                 'years_played']


discr_data = fil_data[focus_columns]



# Preproccessing for  feature scaling to ouput range of 0 to 1 for target variables
scale_columns = ['completions', 'passing_tds', 'attempts', 'interceptions', 
                 'sacks', 'passing_air_yards', 'rushing_yards', 'rushing_tds', 'rushing_first_downs', 'comp_percentage', 'offense_snaps']

features = fil_data[scale_columns]

scaler = MinMaxScaler()

scaled_features = scaler.fit_transform(features)

fil_data[scale_columns] = scaled_features  

discr_data = fil_data[focus_columns]


discr_data
```

Question 3:

```python
Linear Regression Model
```

```python
# Partitioning data via test and train sets
x =  discr_data.select_dtypes(include=['float64', 'int64']).drop('completions', axis=1)
y = discr_data['completions']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=1)

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

y_linear_predict = linear_model.predict(x_test)


x_test['Linear_Model_Completions_Predict'] = y_linear_predict
discr_data = pd.merge(discr_data, x_test[['Linear_Model_Completions_Predict']], left_index=True, right_index=True, how='left')

discr_data.insert(6, 'Linear_Model_Completions_Predict', discr_data.pop('Linear_Model_Completions_Predict'))

discr_data

```

```python
linear_eval = mean_squared_error(y_test, y_linear_predict)
print(linear_eval)
```

```python
Random Forest Regressor Model
```

```python
# Partitioning data via test and train sets
x =  discr_data.select_dtypes(include=['float64', 'int64']).drop('completions', axis=1)
y = discr_data['completions']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=1)

ran_reg_model = RandomForestRegressor(n_estimators=100, random_state=1)
ran_reg_model.fit(x_train, y_train)

y_ran_reg_predict = ran_reg_model.predict(x_test)

x_test['Random_Forest_Completions_Predict'] = y_ran_reg_predict

discr_data = pd.merge(discr_data, x_test[['Random_Forest_Completions_Predict']], left_index=True, right_index=True, how='left')

discr_data.insert(7, 'Random_Forest_Completions_Predict', discr_data.pop('Random_Forest_Completions_Predict'))

discr_data
```

```python
ran_reg_eval = mean_squared_error(y_test, y_ran_reg_predict)
print(ran_reg_eval)
```

I opted to illustrate the results from two models: Linear Regression Model and a Random Forest Regressor Model. The features of both models included all variables except the sorted categorical variable, specifically designed to make both prediction models exclusive to Russell Wilson.

I chose this approach because I did not expect to see a linear correlation between the defined features of the dataset and the target (completions). Nevertheless, I felt it would be valuable to illustrate the differences in output between the two models.

I chose to evaluate the Random Forest Regressor Model due to multiple conflicting variables fundamentally associated with this data and Russell Wison's total completions each game. The most significant factors include:

1. The team he is playing against and the strength of their defense.
  
2. The consistency of his wide receivers in catching his passes.

3. The variation in how often Russell throws check-down passes for short gains compared to longer throws for larger gains.

4. The effect of stadium noise on Russel's ability to remain remain composed and perform each game.

While an attempt could be made to directly account for these variables through feature engineering, acknowledging these unaccounted-for variables in the Linear Regression model serves as justification for the Random Forest Regressor.

Despite these logical justifications for my doubts about the Linear Regression Model compared to the Random Forest Regressor model, I also employed a mean squared error test for both models. Based on the result of these tests, I discovered that the Linear Regression Model would be more suitable, with a mean squared error value of 0.005283226 less than that of the Random Forest Regressor Model.
